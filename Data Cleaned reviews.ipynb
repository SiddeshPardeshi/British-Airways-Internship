# Data Collection :

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#create an empty list to collect all reviews
reviews  = []

#create an empty list to collect rating stars
stars = []

#create an empty list to collect country the reviewer is from
country = []
```

```python
base_url = "https://www.airlinequality.com/airline-reviews/british-airways"
pages = 20
page_size = 150
for i in range(1, pages+1):
    page = requests.get(f"https://www.airlinequality.com/airline-reviews/british-airways/page/{i}/?sortby=post_date%3ADesc&pagesize=150")
    
    soup = BeautifulSoup(page.content, "html.parser")
    
    for item in soup.find_all("div", class_="text_content"):
        reviews.append(item.text)
    
    for item in soup.find_all("div", class_ = "rating-10"):
        try:
            stars.append(item.span.text)
        except:
            print(f"Error on page {i}")
            stars.append("None")
            
        
    #country
    for item in soup.find_all("h3"):
        country.append(item.span.next_sibling.text.strip(" ()"))
```

```python
#check the length of total reviews extracted
len(reviews)
```

```python
len(country)
```

```python
stars = stars[:3000]
```

```python
len(stars)
```

```python
#create  a dataframe from these collected lists of data

data = pd.DataFrame({"reviews":reviews,"stars": stars, "country": country})
```

```python
data.to_csv('BA_reviews.csv', encoding = 'utf-8')

import os
os.getcwd()
```

# Data Cleaning :

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('BA_reviews.csv', index_col = 0)
```

```python
data.head()
```

```python
#We will also create a new column which shows if the person is Verified or not

data['Verified'] = data.reviews.str.contains('Trip Verified')
```

```python
data['Verified']
```

```python
#Cleaning stars column
data['stars'].unique()
```

```python
#removing \n and \t from ratings
data.stars = data.stars.str.strip("\n\t\t\t\t\t\t\t\t\t\t\t\t\t")
```

```python
data['stars'].value_counts()
```

```python
#dropping the row where the value of rating is None

data.drop(data[data.stars == "None"].index, axis=0, inplace=True)
```

```python
data['stars'].unique()
```

```python
data.head()
```

```python
#removing the parts before | in the reviews column
data.reviews = data.reviews.str.split('|',expand=True)[1]
```

```python
data.head()
```

```python
Reviews_df = data.reviews
Reviews_df
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import re
lemma = WordNetLemmatizer()

#creating an empty list to collect cleaned data 
Reviews = []
for rev in Reviews_df:
    rev = re.sub('[^a-zA-Z]',' ', str(rev))
    rev = rev.lower()
    rev = rev.split()
    rev = [lemma.lemmatize(word) for word in rev if word not in set(stopwords.words("english"))]
    rev = " ".join(rev)
    Reviews.append(str(rev))
```

```python
data['Reviews'] = Reviews
```

```python
data.head()
```

```python
#Checking null values

data.isna().sum()
```

```python
data['country'] = data['country'].fillna(data['country'].mode()[0])
```

```python
data['country'].isna().sum()
```

```python
data['country'].isnull().value_counts()
```

```python
data.shape
```

```python
#dropping reviews column
data.drop('reviews', axis = 1, inplace = True)
```

```python
data.head()
```

```python
#resetting the index
data.reset_index(drop=True)
```

```python
data.to_csv('Ba_reviews_cleaned.csv', encoding = 'utf-8')

import os
os.getcwd()
```
